# Install python packages in blender using Windows
Tested with Blender 4.2 <br>
Johannes Beck, 30.10.2024 <br><br>
Note, that the Blender version 4.2 comes with python 3.11.7 and pip installed by default.

<ol>
    <li> Open Blender </li>
    <li> Go to scripting tab </li>
    <li> Go to the interactive python console and get the location of the python installation, that comes with Blender. <br> Therefore, run this: </li>
</ol>

```python
import sys
print(sys.exec_prefix)
```

<ol start="4">
    <li> Run the console (cmd) as administrator </li>
    <li> Change the directory to the on
    e pip install. Specify the site-packages path of the python installation, that blender uses. <br>
    In the shown case for the installation of the package "scipy" type: <br>
    .\bin\python.exe -m pip install scipy --target="C:\Program Files\Blender Foundation\Blender 4.2\4.2\python\lib\site-packages"
</ol>

<ol start="7">
    <li> Restart Blender (might be necessary)</li>
    <li> Test the newly installed packages </li>
</ol>
